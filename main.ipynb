```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Stock Market Dataset.csv')
```

```python
total_length = len(df)
print(f'Total Length of the Dataset: {total_length}')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Natural_Gas_Price'], bins=20, kde=True)
plt.title('Distribution of Natural Gas Price')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
numeric_df = df.drop(columns=['Date'])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Natural_Gas_Price', y='Natural_Gas_Vol.', data=df)
plt.title('Scatter Plot: Natural Gas Price vs. Natural Gas Vol.')
plt.xlabel('Column1')
plt.ylabel('Column2')
plt.show()
```
